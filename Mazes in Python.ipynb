**Task 1; Representing the maze in python.** 
**Explanation**

In this task, the rows increase as you move downward, from 0 to 4 while the columns increase when you move right from 0 to 4. 
The structure used is lists within a lists which are also the rows. Every character represents a cell : 'S' is the beginning, 'G' is the goal, '.' is for free space you can move through and '#' is a wall.



```python
maze = [['s','.','#','.','.'],
       ['.','#','.','#','.'],
       ['.','#','.','.','.'],
       ['.','.','#','#','.'],
       ['.','.','.','.','G']]
start = (0,0)
goal = (4,4)

rows = len(maze)
cols = len(maze[0])
```

**Task 2; Discover paths**

**a. Represent a simple maze as a grid.**
**b. DFS/BFS: Traverse the maze to find any path from S to G.** 
**c. Report the path length and nodes expanded (consider, the cost to be 1 per step).**

```python
def get_neighbors(x, y, maze):
    rows = len(maze)
    cols = len(maze[0])
    moves = [(1,0), (-1,0), (0,1), (0,-1)]
    
    neighbors = []
    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        if 0 <= nx < rows and 0 <= ny < cols:
            if maze[nx][ny] != '#':
                neighbors.append((nx, ny))
    return neighbors
```

```python
from collections import deque
queue = deque([start])
visited = set()
parent = {start: None}
nodes_expanded = 0

while queue:
    x, y = queue.popleft()
    nodes_expanded += 1
    if (x, y) == goal:
        break
        
    moves = [(1,0), (-1,0), (0,1), (0,-1)]
    for dx, dy in moves:
        nx = x + dx
        ny = y + dy
        if 0 <= nx < 5 and 0 <= ny < 5:
            if maze[nx][ny] != "#" and (nx, ny) not in visited and (nx, ny) not in parent:
                parent[(nx, ny)] = (x, y)
                queue.append((nx, ny))
    visited.add((x, y))
# RECONSTRUCT PATH
path = []
node = goal
while node is not None:
    path.append(node)
    node = parent.get(node)
path.reverse()
print("BFS Path:", path)
print("Path length:", len(path))
print("Nodes expanded:", nodes_expanded)
```

```python
def heuristic(node, goal):
    x1, y1 = node
    x2, y2 = goal
    return abs(x1 - x2) + abs(y1 - y2)
```

**Task 4:Greedy best first search**

```python
import heapq

def greedy_best_first(maze, start, goal):
    pq = []
    heapq.heappush(pq, (heuristic(start, goal), start))
    parent = {start: None}
    visited = set()
    nodes_expanded = 0

    while pq:
        _, node = heapq.heappop(pq)
        nodes_expanded += 1

        if node == goal:
            break

        visited.add(node)

        for n in get_neighbors(maze, node):
            if n not in visited and n not in parent:
                parent[n] = node
                heapq.heappush(pq, (heuristic(n, goal), n))

    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = parent[cur]
    path.reverse()
    return path, nodes_expanded
```

**Task 4:A*.**

```python
def a_star(maze, start, goal):
    pq = []
    heapq.heappush(pq, (0, start))
    parent = {start: None}
    g = {start: 0}
    nodes_expanded = 0

    while pq:
        _, node = heapq.heappop(pq)
        nodes_expanded += 1

        if node == goal:
            break

        for n in get_neighbors(maze, node):
            cost = g[node] + 1
            if n not in g or cost < g[n]:
                g[n] = cost
                f = cost + heuristic(n, goal)
                parent[n] = node
                heapq.heappush(pq, (f, n))

    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = parent[cur]
    path.reverse()
    return path, nodes_expanded

gbfs_path, gbfs_expanded = greedy_best_first(maze, start, goal)
astar_path, astar_expanded = a_star(maze, start, goal)

print("Greedy path length:", len(gbfs_path))
print("Greedy nodes expanded:", gbfs_expanded)

print("A* path length:", len(astar_path))
print("A* nodes expanded:", astar_expanded)
```
