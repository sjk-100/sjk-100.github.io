```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

from sklearn.svm import SVC

from sklearn.neighbors import KNeighborsClassifier


data = pd.read_csv("C:/Users/msjka/Downloads/Data.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data =  data.dropna()
data.info()
```

```python
print(data.shape)
```

```python
data.head()
```

```python
#convert LoanApprovalStatus into numerical values
data['LoanApprovalStatus'] = data ['LoanApprovalStatus'].astype('category')
data['LoanApprovalStatus'] = data ['LoanApprovalStatus'].cat.codes
data
```

Split data into dependent and independent variables

```python
x = data.drop(['LoanApprovalStatus'],axis = 1)
x.head()
```

```python
y = data [['LoanApprovalStatus']]
y.head()
```

split data into training and testing data

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size = 0.3,random_state = 21)
```

```python
x_train
```

```python
x_test
```

```python
#Standardizing the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test) 
x_train_scaled
```

Models : Logistic Regression, Support Vector Machines (SVM), and K-Nearest Neighbor (KNN)

1. Logistic Regression

```python
#Logistic regression
Model1 = LogisticRegression(random_state = 21)
Model1 = Model1.fit(x_train_scaled,y_train)
y_pred = Model1.predict(x_test_scaled)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy Score is", round(accuracy*100),"%")
```

```python
print(classification_report(y_test,y_pred))
```

2. Support Vector Machines(SVM)

```python
#Support Vector Machines
Model2 = SVC(kernel='linear')
Model2 = Model2.fit(x_train_scaled,y_train)
y_pred = Model2.predict(x_test_scaled)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy Score is", round(accuracy*100),"%")
```

```python
print(classification_report(y_test,y_pred))
```

3. K-Nearest Neighbor (KNN)

```python
Model3 = KNeighborsClassifier(n_neighbors=5)
Model3 = Model3.fit(x_train_scaled,y_train)
y_pred = Model3.predict(x_test_scaled)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy Score is", round(accuracy*100),"%")
```

```python
print(classification_report(y_test,y_pred))
```
